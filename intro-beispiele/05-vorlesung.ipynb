# Vorlesung 05 - Dependenzgrammatik

```python
import nltk
from nltk import DependencyGraph
from nltk import NonprojectiveDependencyParser
```

```python
sent = "sie fuhren gestern ein sehr schnelles Auto"

grammar = nltk.DependencyGrammar.fromstring("""
'fuhren' -> 'sie' | 'Auto' | 'gestern'
'Auto' -> 'ein' | 'schnelles'
'schnelles' -> 'sehr'
""")

parser = nltk.ProjectiveDependencyParser(grammar)
for tree in parser.parse(sent.split()):
    print(tree)
    tree.pretty_print(unicodelines=True)
```

```python
# nicht-projektive Satzstruktur - kein Parse mit ProjectiveDependencyParser

sent = "sie fuhren ein Auto gestern ein schnelles"

parser = nltk.ProjectiveDependencyParser(grammar)
for tree in parser.parse(sent.split()):
    tree.pretty_print(unicodelines=True)
```

```python
# aber Parsing möglich mit NonprojectiveDependencyParser

dp = NonprojectiveDependencyParser(grammar)
g, = dp.parse(sent.split())

print(g.tree())
g.tree().pretty_print(unicodelines=True)
```
