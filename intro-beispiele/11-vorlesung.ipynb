# Vorlesung 11 - Grammatikinduktion

```python
import nltk
from nltk import Tree
```

```python
#small treebank:
treestrings = [
"""(S
  (NP (PRON er))
  (VP (V schläft)))
""",

"""(S
  (NP (PRON sie))
  (VP
    (V sieht)
    (NP
      (DET eine)
      (N Katze))))
"""
]

trees = [Tree.fromstring(treestring) for treestring in treestrings]

[tree.pretty_print(unicodelines=True) for tree in trees]
```

```python
#grammar induction:
productions = []
S = nltk.Nonterminal('S')

for tree in trees:
    productions += tree.productions()

grammar = nltk.induce_pcfg(S, productions)
for production in grammar.productions():
    print(production)    
```

```python
#parse new sentence with induced grammar:
parser = nltk.ViterbiParser(grammar)

sent = "eine Katze schläft"

for tree in parser.parse(sent.split()):
    print(tree)
    tree.pretty_print(unicodelines=True)
```
