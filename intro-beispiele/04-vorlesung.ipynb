# Vorlesung 04 - Parsingalgorithmen

```python
import nltk
```

```python
grammar = nltk.CFG.fromstring("""
    S -> NP VP
    NP -> PROPN
    VP -> V NP
    PROPN -> "Maria" | "Moritz"
    V -> "sieht"
""")
```

```python
sent = "Maria sieht Moritz"

parser = nltk.RecursiveDescentParser(grammar, trace=3)

for tree in parser.parse(sent.split()):
    tree.pretty_print(unicodelines=True)
```

```python
sent = "Maria sieht Moritz"

parser = nltk.ShiftReduceParser(grammar, trace=2)

for tree in parser.parse(sent.split()):
    tree.pretty_print(unicodelines=True)
```

```python
sent = "Maria sieht Moritz"

parser = nltk.EarleyChartParser(grammar, trace=2)

for tree in parser.parse(sent.split()):
    print(tree)
    tree.pretty_print(unicodelines=True)
```

```python
sent = "Maria sieht Moritz"

parser = nltk.ChartParser(grammar, trace=2)

for tree in parser.parse(sent.split()):
    print(tree)
    tree.pretty_print(unicodelines=True)
```
