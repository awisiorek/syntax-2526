# Vorlesung 03 - X-Bar-Struktur

```python
import nltk
from nltk.parse.generate import generate
```

```python
sent = "sie sind wieder ein sehr sehr schnelles Auto gefahren gestern"

grammar = nltk.CFG.fromstring(
"""
    S -> NP VP
    NP -> PRON

# X-Bar-Regeln NP:
    NP -> DET NOM
    NOM -> ADJP NOM
    NOM -> N

# X-Bar-Regeln ADJP:
    ADJP -> ADJ_BAR
    ADJ_BAR -> PRT ADJ_BAR 
    ADJ_BAR -> ADJ

# X-Bar-Regeln VP:
    VP -> AUX VERBAL
    VERBAL -> ADV VERBAL |  VERBAL ADV
    VERBAL -> NP V

#lexikalische Regeln:
    PRON -> 'sie'
    AUX -> 'sind'
    ADV -> 'gestern' | 'wieder'
    V -> 'gefahren'
    DET -> 'ein'
    PRT -> 'sehr'
    ADJ -> 'schnelles' 
    N -> 'Auto'
""")

parser = nltk.ChartParser(grammar)

for tree in parser.parse(sent.split()):
    tree.pretty_print(unicodelines=True)
```

```python
for sentence in generate(grammar, depth=7):
    print(' '.join(sentence))
```
