# Vorlesung 12 - Partielles Parsing

```python
import nltk
from nltk.tag.util import str2tuple
```

```python
sent_tagged = "sie/PRON sind/AUX wieder/ADV ein/DET sehr/PRT sehr/PRT schnelles/ADJ Auto/N gefahren/V gestern/ADV"
sent_pos = [str2tuple(tok) for tok in sent_tagged.split()]
```

```python
regex = r"""
ADJP:
    {<PRT|ADV>*<ADJ>+}

NP:
    {<DET>?<ADJP>*<N>+}
"""
```

```python
chunk_parser = nltk.RegexpParser(regex)
```

```python
chunk_parser.parse(sent_pos)
```
