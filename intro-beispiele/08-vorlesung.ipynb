# Vorlesung 08 - Grammatische Merkmale

```python
import nltk
from nltk.featstruct import FeatStruct
```

### Beispiel 1: keine Subsumption, aber unifizierbar

```python
f0 = FeatStruct("[A=1, B=2]")
print(f0)
```

```python
f1 = FeatStruct("[A=1, C=3]")
print(f1)
```

```python
print(f0.unify(f1))
```

```python
f0.subsumes(f1)
```

### Beispiel 2: keine Subsumption, nicht unifizierbar (widersprüchliche Merkmalswerte)

```python
f0 = FeatStruct("[A=1, B=2]")
print(f0)
```

```python
f1 = FeatStruct("[A=1, B=3]")
print(f1)
```

```python
print(f0.unify(f1))
```

```python
f0.subsumes(f1)
```

### Beispiel 3: Subsumptionsrelation, also unifizierbar (ergibt spezifischere Struktur)

```python
f0 = FeatStruct("[A=1]")
print(f0)
```

```python
f1 = FeatStruct("[A=1, B=3]")
print(f1)
```

```python
print(f0.unify(f1))
```

```python
f0.subsumes(f1)
```
